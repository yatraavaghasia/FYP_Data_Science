### Importing Data

```python
import pandas as pd
trump_data_path = '/content/drive/MyDrive/FYP_SEM_VII/Datasets/hashtag_donaldtrump.csv'
trump_df = pd.read_csv(trump_data_path,chunksize=5000)
trump_df = trump_df.get_chunk()
trump_df.head(5000)
```

### Filtering out Non-English Tweets
Detecting the language of the tweets using the roberta-base language detection model that employs transformers, and then removing non-english language tweets from the dataset.

https://huggingface.co/papluca/xlm-roberta-base-language-detection

```python
 from transformers import pipeline

# Load the language detection pipeline
lang_detector = pipeline("text-classification", model="papluca/xlm-roberta-base-language-detection")

# Function to detect language
def detect_language(tweet):
    return lang_detector(tweet)[0]['label']

# Apply language detection to each tweet
trump_df['language'] = trump_df['tweet'].apply(detect_language)

# Display the first few rows with detected language
print(trump_df[['tweet_id', 'tweet', 'language']].head())
```

### No. of English language tweets out of 5000: 3836

```python
# Get the total number of tweets for each unique language
language_counts = trump_df['language'].value_counts()

# Display the counts
print(language_counts)

# Filter the DataFrame to keep only tweets in English
english_tweets_df = trump_df[trump_df['language'] == 'en']

# Display the first few rows of the filtered DataFrame
print(english_tweets_df.head())
```

### Preprocessing and cleaning the tweets
These are the functions performed:


*   Convert text to lowercase
*   Remove URLs
*   Remove user mentions and hashtags
*   Remove punctuations
*   Removing extra spaces


```python
import re  # Regular expression operations library
import string  # String operations library
import pandas as pd  # Pandas library for data manipulation

# Sample preprocessing function
def preprocess_tweet(tweet):
    # Convert all characters in the tweet to lowercase
    tweet = tweet.lower()

    # Remove URLs starting with 'http', 'www', or 'https' by replacing them with an empty string
    tweet = re.sub(r"http\S+|www\S+|https\S+", '', tweet, flags=re.MULTILINE)

    # Remove user mentions (e.g., '@username') and hashtags (e.g., '#hashtag') by replacing them with an empty string
    tweet = re.sub(r'\@\w+|\#', '', tweet)

    # Remove punctuation marks by translating each punctuation character to None
    tweet = tweet.translate(str.maketrans('', '', string.punctuation))

    # Remove any single characters surrounded by spaces (e.g., ' a ') by replacing them with a single space
    tweet = re.sub(r'\s+[a-zA-Z]\s+', ' ', tweet)

    # Remove extra spaces by replacing multiple spaces with a single space and strip leading/trailing spaces
    tweet = re.sub(r'\s+', ' ', tweet).strip()

    # Return the cleaned tweet
    return tweet

# Apply the preprocessing function to each tweet in the 'tweet' column of the DataFrame 'df'
# and store the cleaned tweets in a new column 'cleaned_tweet'
df = pd.DataFrame({
    'tweet_id': english_tweets_df['tweet_id'],  # Assuming the original DataFrame has a 'tweet_id' column
    'cleaned_tweet': english_tweets_df['tweet'].apply(preprocess_tweet)
})
print(df.head())
```

## Imported pre-trained model and predicted sentiments
Used the sentiment-analysis pipeline to predict sentiments of the tweets. The sentiment-analysis pipeline from the Hugging Face transformers library uses a pre-trained model to classify the sentiment of text. In this case, the pre-trained model is `distilbert/distilbert-base-uncased-finetuned-sst-2-english`

Binary-class classification: POSITIVE/NEGATIVE

```python
from transformers import pipeline
import pandas as pd

# Load a pre-trained sentiment analysis model
sentiment_model = pipeline("sentiment-analysis")

# Assuming your DataFrame is named 'df' and has a 'cleaned_tweet' column
# Apply sentiment analysis to each cleaned tweet
def predict_sentiment(tweet):
    result = sentiment_model(tweet)[0]
    return result['label'], result['score']

# Apply the sentiment prediction function to each tweet
df[['sentiment_label', 'sentiment_score']] = df['cleaned_tweet'].apply(lambda x: pd.Series(predict_sentiment(x)))

# Display the DataFrame with predictions
print(df.head())
```

```python
# Print the model name and other details
print(f"Model used: {sentiment_model.model.config._name_or_path}")
print(f"Model configuration: {sentiment_model.model.config}")
```

```python
# Randomly select two tweets
tweets_to_inspect = df.sample(n=2, random_state=42)  # Set random_state for reproducibility if needed

# Display the full text of the tweets and their associated predicted sentiments
for index, row in tweets_to_inspect.iterrows():
    print(f"Tweet ID: {row['tweet_id']}")
    print(f"Tweet: {row['cleaned_tweet']}")
    print(f"Predicted Sentiment: {row['sentiment_label']} with confidence {row['sentiment_score']:.2f}")
    print("-" * 40)
```

## Import the pre-trained model and predict the sentiments
The `pipeline("sentiment-analysis")` from the Hugging Face `transformers` library automatically loads a pre-trained model for sentiment classification, such as `finiteautomata/bertweet-base-sentiment-analysis`, which is fine-tuned for analyzing tweets. It handles tokenization, padding, and attention masks, converting input text into a format the model can process. The model performs inference to predict sentiment and provides labels and confidence scores. The pipeline abstracts these steps, allowing you to easily apply sentiment analysis to your data with minimal code.

```python
from transformers import pipeline
import pandas as pd

# Load the pre-trained sentiment analysis model specific for tweets
specific_model = pipeline(model="finiteautomata/bertweet-base-sentiment-analysis")

# Function to predict sentiment for a single tweet
def predict_sentiment(tweet):
    result = specific_model(tweet)[0]
    return result['label'], result['score']

# Apply sentiment analysis to each cleaned tweet in the DataFrame
df[['sentiment_label', 'sentiment_score']] = df['cleaned_tweet'].apply(lambda x: pd.Series(predict_sentiment(x)))

# Display the DataFrame with sentiment predictions
print(df.head())
```

```python
# Print the model name and other details
print(f"Model used: {specific_model.model.config._name_or_path}")
print(f"Model configuration: {specific_model.model.config}")
```

```python
# Randomly select two tweets
tweets_to_inspect = df.sample(n=2, random_state=42)  # Set random_state for reproducibility if needed

# Display the full text of the tweets and their associated predicted sentiments
for index, row in tweets_to_inspect.iterrows():
    print(f"Tweet ID: {row['tweet_id']}")
    print(f"Tweet: {row['cleaned_tweet']}")
    print(f"Predicted Sentiment: {row['sentiment_label']} with confidence {row['sentiment_score']:.2f}")
    print("-" * 40)
```
